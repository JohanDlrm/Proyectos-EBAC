```python

# Obtener los años únicos ordenados
años = sorted(dfxgenero['Year'].unique())

# Crear subplots: ajusta según cuántos años haya (16 → 4x4)
n = len(años)
ncols = 4
nrows = (n + ncols - 1) // ncols  # redondear hacia arriba

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20, nrows * 5))
axes = axes.flatten()

# Crear un gráfico de pastel por año
for i, año in enumerate(años):
    ax = axes[i]
    data = dfxgenero[dfxgenero['Year'] == año]
    
    ax.pie(
        data['Global_Sales'],
        labels=data['Genre'],
        autopct='%1.1f%%',
        startangle=90,
        textprops={'fontsize': 8}
    )
    ax.set_title(f'Año {int(año)}', fontsize=12)

# Ocultar subplots vacíos (si hay más ejes que años)
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.suptitle('Proporción de Ventas Globales por Género de Videojuego (2000–2015)', fontsize=18)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Deja espacio para el título
plt.savefig("ventas_por_genero_por_year.png", dpi=300, bbox_inches='tight')
plt.show()

```

```python
tabla_por_juego = pd.pivot_table(df4, values = 'Global_Sales', index = ['Year'],
               columns = ['Name'], aggfunc = sum).fillna(0).reset_index()
tabla_por_juego.T
```

```python
# 1. Calcular crecimiento porcentual año a año para cada juego (columna)
tabla_pct_crecimiento = tabla_por_juego.set_index('Year').pct_change() * 100

# 2. Reemplazar NaN (por el primer año sin valor previo) por 0 o dejarlo
tabla_pct_crecimiento = tabla_pct_crecimiento.fillna(0)

# 3. Para cada juego, obtener el máximo crecimiento y el mínimo crecimiento
max_crecimiento = tabla_pct_crecimiento.max()
min_crecimiento = tabla_pct_crecimiento.min()

# 4. Juego con mayor crecimiento
juego_max = max_crecimiento.idxmax()

# 5. Juego con peor crecimiento 
juego_min = min_crecimiento.idxmin()

print(f"El juego con mayor crecimiento anual en ventas es '{juego_max}' ")
print(f"El juego con peor crecimiento anual en ventas es '{juego_min}' ")
```

```python
#qué género ha tenido el mayor y menor crecimiento.
dfgenero = (
    df2.groupby(['Year', 'Genre'], as_index=False)['Global_Sales']
    .sum()
    .sort_values(['Year', 'Global_Sales'], ascending=[True, False])
)

dfgenero
```

```python
pivot = dfgenero.pivot(index='Year', columns='Genre', values='Global_Sales').fillna(0)
pivot.head(3)
```

```python
diff_abs = pivot.diff()
diff_abs.head(3)
```

```python
pct_change = pivot.pct_change() * 100
pct_change.head(3)
```

```python
#Ver el crecimiento por diferencia absoluta
crecimiento_total_abs = pivot.iloc[-1] - pivot.iloc[0]
crecimiento_total_abs.sort_values(ascending=False)
```

```python
#Ver el crecimiento porcenutal acumulado (promedio anual)
crecimiento_porcentual_promedio = pct_change.mean()
crecimiento_porcentual_promedio.sort_values(ascending=False)
```

```python
#Asignar color según tendencia
colors = ['skyblue' if val >= 0 else 'salmon' for val in crecimiento_total_abs]

#Graficar
plt.figure(figsize=(12, 6))
crecimiento_total_abs.plot(kind='bar', color=colors)
plt.title('Crecimiento Absoluto Total por Género (Ventas Globales)')
plt.xlabel('Género')
plt.ylabel('Crecimiento en Ventas (millones)')
plt.xticks(rotation=45)
plt.tight_layout()

# Guardar imagen
plt.savefig("crecimiento_absoluto_por_genero_coloreado.png", dpi=300)
plt.show()
```

```python
#  plataforma con mayor y menor crecimiento.
dfplatform = (
    df2.groupby(['Year', 'Platform'], as_index=False)['Global_Sales']
    .sum()
    .sort_values(['Year', 'Global_Sales'], ascending=[True, False])
)

pivot_platform = dfplatform.pivot(index='Year', columns='Platform', values='Global_Sales').fillna(0)
pivot_platform
```

```python
diff_abs_platform = pivot_platform.diff()
diff_abs_platform
```

```python
#Ver el crecimiento por diferencia absoluta
crecimiento_total_abs_platform = diff_abs_platform.sum().sort_values(ascending=False)
crecimiento_total_abs_platform
```

```python
# Colores según crecimiento positivo o negativo
colors = ['skyblue' if val >= 0 else 'salmon' for val in crecimiento_total_abs_platform]

# Gráfica
plt.figure(figsize=(14, 6))
bars = plt.bar(crecimiento_total_abs_platform.index, crecimiento_total_abs_platform.values, color=colors)
plt.title('Crecimiento Absoluto Total por Plataforma (Ventas Globales)')
plt.xlabel('Plataforma')
plt.ylabel('Crecimiento en Ventas (millones)')
plt.xticks(rotation=45)

# Etiquetas encima de cada barra
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.2f}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 5),
                 textcoords="offset points",
                 ha='center', va='bottom',
                 fontsize=9)

plt.tight_layout()
plt.savefig("crecimiento_absoluto_por_plataforma.png", dpi=300)
plt.show()
```
